## Loading raw wine production data

In this notebook, we will:

1. Load the raw wine production data per municipality (Excel file).
2. Inspect the first few rows to verify correctness.
3. Save the dataset as CSV for easier processing in later steps.


```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Caminho do ficheiro
file_path = r"C:\Users\islec\climate-wine-project\data_raw\wine_production_raw.xlsx"
output_path = r"C:\Users\islec\climate-wine-project\data_clean\wine_production_clean_.csv"
os.makedirs(os.path.dirname(output_path), exist_ok=True)

# Ler a aba 22
df = pd.read_excel(file_path, sheet_name=22, header=None)

# Definir nomes das colunas
df.columns = [
    'Distrito', 'Concelho',
    'Vinho_Tinto', 'Vinho_Rosado', 'Vinho_Branco', 'Vinho_Total',
    'IndCasta_Tinto', 'IndCasta_Rosado', 'IndCasta_Branco', 'IndCasta_Total',
    'IGP_Tinto', 'IGP_Rosado', 'IGP_Branco', 'IGP_Total',
    'LicorosoDOP_Tinto', 'LicorosoDOP_Rosado', 'LicorosoDOP_Branco', 'LicorosoDOP_Total',
    'DOP_Tinto', 'DOP_Rosado', 'DOP_Branco', 'DOP_Total',
    'TotalGeral_Tinto', 'TotalGeral_Rosado', 'TotalGeral_Branco', 'TotalGeral_Total'
]

# Selecionar apenas Totais Gerais
df_clean = df[['Distrito', 'Concelho', 'TotalGeral_Tinto', 'TotalGeral_Rosado',
               'TotalGeral_Branco', 'TotalGeral_Total']].copy()

# Remover linhas vazias ou de cabeçalhos
df_clean = df_clean.dropna(how='all')
df_clean = df_clean[~df_clean['Distrito'].astype(str).str.contains('Produção|Total|Distrito', na=False)]

# Preencher o nome do distrito
df_clean['Distrito'] = df_clean['Distrito'].ffill()

# Converter colunas numéricas
for col in ['TotalGeral_Tinto', 'TotalGeral_Rosado', 'TotalGeral_Branco', 'TotalGeral_Total']:
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce').fillna(0)

# Remover possíveis linhas iniciais nulas
df_clean = df_clean[df_clean['Concelho'].notna()].reset_index(drop=True)

# 💡 Calcular total por distrito
df_distrito = df_clean.groupby('Distrito', as_index=False)[
    ['TotalGeral_Tinto', 'TotalGeral_Rosado', 'TotalGeral_Branco', 'TotalGeral_Total']
].sum()

# Salvar CSV final
df_clean.to_csv(output_path, index=False)

print("✅ CSV final gerado com sucesso!")
print(df_clean.head(10))
print("\nResumo por distrito:")
print(df_distrito.head(10))
```

```python
# 📁 Criar CSV apenas com o total por distrito

# Arredondar os valores
df_distrito = df_distrito.round(0)

# Caminho do novo ficheiro
output_districts = r"C:\Users\islec\climate-wine-project\data_clean\wine_production_by_district.csv"
os.makedirs(os.path.dirname(output_districts), exist_ok=True)

# Salvar CSV só com totais por distrito
df_distrito.to_csv(output_districts, index=False)

print("✅ CSV com totais por distrito criado com sucesso!")
print(df_distrito.head(10))
print(f"\n📂 Ficheiro salvo em: {output_districts}")
```

```python

# 📊 Criar gráfico de barras horizontais
plt.figure(figsize=(10, 7))
df_plot = df_distrito.sort_values('TotalGeral_Total', ascending=True)

bars = plt.barh(df_plot['Distrito'], df_plot['TotalGeral_Total'],
                color=['#6a1b1a' if i >= len(df_plot)-5 else '#b56576' for i in range(len(df_plot))])

plt.title("Produção Total de Vinho por Distrito (hl)", fontsize=14, fontweight='bold')
plt.xlabel("Hectolitros (hl)")
plt.ylabel("Distrito")
plt.grid(axis='x', linestyle='--', alpha=0.4)

# Adicionar valores nas barras
for bar in bars:
    plt.text(bar.get_width() + 100, bar.get_y() + bar.get_height()/2,
             f"{bar.get_width():,.0f}", va='center', fontsize=9)

# Ajustar layout e salvar
plt.tight_layout()
viz_path = r"C:\Users\islec\climate-wine-project\data_viz\wine_total_by_district.png"
os.makedirs(os.path.dirname(viz_path), exist_ok=True)
plt.savefig(viz_path, dpi=300, bbox_inches='tight')

plt.show()

print(f"\n📊 Gráfico salvo em: {viz_path}")
```
